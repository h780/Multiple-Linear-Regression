# Fitting the Model on training and testing sets

```python
columns = df.columns
columns
```

```python
#Extracting predictor and response variables
X = df[['DistanceCovered(InKms)', 'Goals', 'ShotsPerGame', 'AgentCharges', 'BMI', 'Cost', 'PreviousClubCost']]

y = df['Score']
```

```python
#Splitting as training and testing set

x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, test_size = 0.25, random_state = 100)
```

```python
#Forcing intercept term, i.e. beta0 and fitting the OLS model on updated x_train

x_train_with_intercept = sm.add_constant(x_train)
lr1 = sm.OLS(y_train, x_train_with_intercept).fit()
lr1.summary()
```

So, our R_squared is great i.e. 95.9% and adjusted R-squared also. The number of observations we have is not similar to number of predictor variables. Hence, adjusted R_squared is not super critical. We will try to eliminate Goals, ShotsPerGame, and AgentCharges as they don't seem to be very meaningful from the T-test prespective. Their t-test probability values points towards not rejecting the null hypothesis.

```python
#Running the model after trimming some variables.

X1 = df[['DistanceCovered(InKms)', 'BMI', 'Cost', 'PreviousClubCost']]

x1_train, x1_test, y1_train, y1_test = train_test_split(X1, y, train_size = 0.75, test_size = 0.25, random_state = 100)

x1_train_with_intercept = sm.add_constant(x1_train)
lr2 = sm.OLS(y1_train, x1_train_with_intercept).fit()
lr2.summary()
```

Our AIC & BIC values are dropped which is a good result. There is no change in R_squared and adjusted R_squared, this is algo acceptable result. All the other tests are also looking promising.

```python
#Throwing in Club in the data by performaing one-hot encoding

#Extracting club from dataframe into a variable
clubs = set(df.Club)
clubs

#Doing one-hot-encoding using get_dummies variable
nominal_features = pd.get_dummies(df['Club'])
nominal_features
```

We can see we have three clubs here in our data, i.e. Chelsea, Liverpool, and Manchester United. All the players in the dataset are assigned with values for the club from they belong.

```python
#Inserting the encoded club variables in the dataframe using concat function from pandas libarary. This only works when we have 
#same number of rows in both the dataframes, and it will join side by side.

df_encoded = pd.concat([df, nominal_features], axis = 1)
df_encoded
```

So, now we have added our encoded variables, CHE, LIV, and MUN to the main dataframe.

```python
#Running the model after adding encoded variables

X2 = df_encoded[['DistanceCovered(InKms)', 'BMI', 'Cost', 'PreviousClubCost', 'CHE', 'LIV', 'MUN']]
X3 = pd.concat([df['Score'], X2], axis=1)

fig = px.scatter_matrix(X3,
                        title = 'Scatter plot for best model dataframe',
                        labels={col:col.replace('DistanceCovered(InKms)','DistCov') for col in X3.columns},
                        width=1000, 
                        height=1200)
fig.show()

x2_train, x2_test, y2_train, y2_test = train_test_split(X2, y, train_size = 0.75, test_size = 0.25, random_state = 100)

x2_train_with_intercept = sm.add_constant(x2_train)
lr3 = sm.OLS(y2_train, x2_train_with_intercept).fit()
lr3.summary()
```

So, we can see that our R_squared and adjusted R_squared values increased after adding Club info, which says that adding Club explains more variance from the data. The AIC & BIC went down, which is good as we need to minimize them. All the test values are good, except the probability of t-test for LIV, which is 0.515 and it somewhat rejects the alternative hypothesis.

```python
#Looking at the testing data

x2_test_with_intercept = sm.add_constant(x2_test)
y2_test_fitted = lr3.predict(x2_test_with_intercept)

plt.scatter(y2_test_fitted, y2_test)
plt.xlabel('Fitted Data')
plt.ylabel('Testing Data')
plt.title('Model Plot')
plt.show()
```

Overall this looks a pretty good model, we should look to get more testing data, if possible.

```python
#Comparing actual versus predicted values

df1 = pd.DataFrame({'Actual': y2_test, 'Predicted': y2_test_fitted})
df1

df1.plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
